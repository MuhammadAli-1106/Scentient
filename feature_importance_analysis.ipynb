```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
importance_df = pd.read_csv("data/feature_importance.csv")
```

```python
top_features = importance_df.head(30)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x="mi_score", y="feature", data=top_features, palette="viridis")
plt.title("Top 30 Features by Mutual Information")
plt.xlabel("Mutual Information Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
